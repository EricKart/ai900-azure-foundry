## Part 1: The 6 Microsoft Responsible AI Principles

Microsoft has established six core principles that guide how AI should be developed and deployed. These are a key topic on the AI-900 exam.

| # | Principle | Description |
|---|-----------|-------------|
| 1 | **Fairness** | AI systems should treat all people fairly and avoid affecting similarly situated groups in different ways |
| 2 | **Reliability & Safety** | AI systems should perform reliably and safely under normal and unexpected conditions |
| 3 | **Privacy & Security** | AI systems should be secure and respect user privacy |
| 4 | **Inclusiveness** | AI systems should empower everyone and engage people, not exclude them |
| 5 | **Transparency** | AI systems should be understandable - people should know how the AI makes decisions |
| 6 | **Accountability** | People should be accountable for AI systems and their impact on society |

### AI-900 Exam Tip
You should be able to **identify** which principle applies to a given scenario. For example:
- A hiring AI that favours one gender? That's a **Fairness** issue.
- An AI chatbot that doesn't explain why it denied a loan? That's a **Transparency** issue.
- A self-driving car AI that crashes in rain? That's a **Reliability & Safety** issue.

## Part 2: Azure AI Content Safety - Live Demo

**Azure AI Content Safety** is a service that detects potentially harmful content in text and images. It checks for four categories:

| Category | What It Detects |
|----------|----------------|
| **Hate** | Content that attacks or discriminates against identity groups |
| **SelfHarm** | Content related to self-harm actions |
| **Sexual** | Sexually explicit content |
| **Violence** | Content describing or promoting violence |

Each category gets a **severity score** from 0 (safe) to 6 (severe).

Let's see it in action with some sample texts!

```python
from azure.ai.contentsafety import ContentSafetyClient
from azure.ai.contentsafety.models import AnalyzeTextOptions
from azure.core.credentials import AzureKeyCredential

# Create the Content Safety client
content_safety_client = ContentSafetyClient(
    endpoint=AI_ENDPOINT,
    credential=AzureKeyCredential(AI_KEY)
)

print("Content Safety client created!")
```

### Test 1: Safe Text

Let's analyze a completely harmless sentence. We expect all severity scores to be 0.

```python
def analyze_text(client, text):
    """Analyze text for harmful content and display results."""
    print(f"Analyzing: \"{text}\"")
    print("-" * 60)
    
    request = AnalyzeTextOptions(text=text)
    response = client.analyze_text(request)
    
    for category in response.categories_analysis:
        severity = category.severity
        # Visual indicator
        if severity == 0:
            indicator = "[SAFE]"
        elif severity <= 2:
            indicator = "[LOW]"
        elif severity <= 4:
            indicator = "[MEDIUM]"
        else:
            indicator = "[HIGH]"
        print(f"  {indicator} {category.category}: severity {severity}/6")
    print()

# Safe text example
analyze_text(content_safety_client, "The weather is beautiful today and the flowers are blooming in the garden.")
```

### Test 2: Mildly Negative Text

This text contains some mild negativity. Let's see how the system classifies it.

```python
analyze_text(content_safety_client, "I am frustrated with this product. It is terrible and I want my money back immediately.")
```

### Test 3: Multiple Texts at Once

Let's analyze several different types of text to see how Content Safety categorizes each one.

```python
sample_texts = [
    "Azure AI provides powerful tools for building intelligent applications.",
    "The movie had some intense action scenes with dramatic special effects.",
    "The medical research team is studying treatments to help patients recover faster.",
    "People from all backgrounds and cultures should be treated with equal respect.",
]

print("Analyzing multiple texts...\n")
for i, text in enumerate(sample_texts, 1):
    print(f"--- Text {i} ---")
    analyze_text(content_safety_client, text)
```

- Severity scores range from **0 to 6** (0 = safe, 6 = severe)
- Four categories: **Hate, SelfHarm, Sexual, Violence**
